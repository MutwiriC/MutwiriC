```python
import matplotlib.pyplot as plt
import numpy as np

#add data
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

#plot scatter plot
plt.scatter(x_data, y_data, c = 'red')
plt.show()
```

```python
#plotting a line graph
year = [2006 + x for x in range(6)]
weight = [80,83,84,85,86, 79]
plt.plot(year, weight, color = "b", lw = 3, linestyle = "--")
plt.show()
```

```python
#working with categorical data
x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [20, 50, 140, 5, 30]

#we use align to make sure the bar starts at the first category
plt.bar(x,y, align = 'edge', width = 0.5, edgecolor = 'red');
plt.show()
```

```python
#plot customization
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55,56,62,61,72,72,73,75]


income_ticks = list(range(50, 81, 2))
plt.plot(years, income)
plt.xlabel('Years')
plt.ylabel("Yearly income in USD")
plt.title("Income of Jon", fontsize = 19, fontname = 'DejaVu Serif')
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])

plt.show()
```

```python
#histogram
# generate 1000 normally distributed random numbers with a mean of 20 and st of 1.5
ages = np.random.normal(20, 1.5, 1000)

#we can adjust the bin size usign bins (can pass a number or a collection)
# we use cumulative to plot cumulative histogram
plt.hist(ages, bins = [ages.min(), 18, 21, ages.max()], cumulative= True)
plt.show()
```

```python
#pie chart
langs = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [20, 50, 14, 5, 18]
#to highlight a specific entry we use explode
explodes = [0,0.01, 0, 0, 0]

#to show percentages we use autopct. And to show how far from the middle we ise pctdistance
#plt.pie(votes, labels=langs, explode = explodes, autopct = '%.2f%%', pctdistance= 1.3)
# we can choose a different start angle
plt.pie(votes, labels=langs, explode = explodes, autopct = '%.2f%%', pctdistance= 1.4, startangle = 90)
plt.legend(loc = 'lower right')
plt.show()
```

```python
votes = [20, 50, 14, 5, 18]
people = ['A', 'B', 'C', 'D', 'E']

plt.pie(votes, labels = None)
plt.legend(labels=people)
plt.show()
```

```python
#styling
import matplotlib.pyplot as plt
from matplotlib import style

#add dack background to the figure
style.use('dark_background')

votes = [20, 50, 14, 5, 18]
people = ['A', 'B', 'C', 'D', 'E']

plt.pie(votes, labels = None)
plt.legend(labels=people)
plt.show()
```

```python
#boxplot
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
plt.show()
```

```python
#plotting multiple figures

x1,y1 = np.random.random(100), np.random.random(100)
x2,y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)

plt.show()

#this creates 2 independent windows
```

```python
#subplotting

x = np.arange(100)

fig, axs =plt.subplots(2,2) #2 rows 2 columns
axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title('Random Function')

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title('log Function')
axs[1,1].set_xlabel("Test")

fig.suptitle("Four Plots")

#to remove overwrap betwen labels

plt.tight_layout()

#plt.show()

#to export graph replace plt.show () with plt.savefig()
#we use dpi to improve graph image quality

plt.savefig('fourplot.png', dpi = 300, transparent =False, bbox_inches = 'tight')

```

```python
#3d plotting and animation

ax = plt.axes(projection = '3d')

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title('3D plotting')
ax.set_xlabel('test')
plt.show()
```

```python
ax = plt.axes(projection = '3d')

x = np.arange(0, 50, 0.1)
y = np.arange(0, 50, 0.1)
z = np.cos(x)

ax.plot(x,y,z)
ax.set_title('3D plotting')
ax.set_xlabel('test')
plt.show()
```

```python
#surface plot
ax = plt.axes(projection = '3d')

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

# Create meshgrid
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Plot
fig = plt.figure()
surf = ax.plot_surface(X, Y, Z, cmap="Spectral");

fig.colorbar(surf)  # add colorbar
ax.set_title('3D plotting')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(azim =0)
plt.show()
```

```python
#animations
from matplotlib import rc
rc('animation', html='jshtml')
import random

heads_tails = [0,0]

for _ in range(100000):
  heads_tails[random.randint(0,1)] +=1
  plt.bar(['Heads', 'Tails'], heads_tails, color = ['red', 'blue'])
  plt.pause(0.001) #pause is used to animate videos
plt.show()
```
